## Analizando trip.csv

```python
trips = pd.read_csv('data/trip.csv',
                    parse_dates=['start_date', 'end_date'], infer_datetime_format=True,
                    low_memory=False)
```

```python
print trips.shape
trips.head(2)
```

```python
for i in trips.columns.values: print i
```

```python
trips['season'] = trips.start_date.dt.month

trips.season = trips.season.replace(1, 'winter')
trips.season = trips.season.replace(2, 'winter')
trips.season = trips.season.replace(3, 'winter')
trips.season = trips.season.replace(4, 'spring')
trips.season = trips.season.replace(5, 'spring')
trips.season = trips.season.replace(6, 'spring')
trips.season = trips.season.replace(7, 'summer')
trips.season = trips.season.replace(8, 'summer')
trips.season = trips.season.replace(9, 'summer')
trips.season = trips.season.replace(10, 'autumn')
trips.season = trips.season.replace(11, 'autumn')
trips.season = trips.season.replace(12, 'autumn')
```

```python
#separamos la fecha de las hora
trips['start_time'] = trips.start_date.dt.time
trips['start_hour'] = trips.start_date.dt.hour
trips['day_of_week'] = trips.start_date.dt.dayofweek
trips['start_date'] = trips.start_date.dt.date

trips['end_time'] = trips.end_date.dt.time
trips['end_hour'] = trips.end_date.dt.hour
trips['end_date'] = trips.end_date.dt.date

trips.head(2)
```

```python
trips.duration.describe()
```

```python
print trips.duration.max()
print trips.duration.min()
```

```python
print list(set(trips.subscription_type.values))
```

```python
print trips.end_time.describe()
print trips.end_time.min()
print trips.start_time.max()
```

```python
# n√∫mero de valores distintos que puede tomar zipcode (demasiados para print)
print len(list(set(trips.zip_code.values)))
```


Debido al alto n√∫mero de trips con duraciones muy chicas, de pocos minutos, teorizamos que estos "trips" no son en verdad alquileres si no bicicletas retiradas para reparaci√≥n o control. Asimismo podemos observar picos de "alquileres" en horas muy extra√±as del d√≠a, que no por casualidad coinciden con los viajes de poca duraci√≥n. Concluimos entonces que estos registros en el set de datos no corresponden a viajes reales.

Por lo tanto filtraremos los registros con duraciones menores a 5 minutos (300 segundos). A su vez, la misma p√°gina aclara que los viajes no puede durar m√°s de 24hrs (86400 segundos) pues entonces se considera la bicicleta como 'robada', por lo que filtramos este tipo de viajes tambi√©n.
Teniendo en cuenta tambi√©n este dato, reemplazamos las columnas start_date y end_date por una √∫nica date correspondiente al d√≠a de comienzo del viaje.

Los zip_codes no parecen brindar informaci√≥n clara. Son demasiados valores y no corresponden con los zip codes de las 5 ciudades de weather y station.

Finalmente, bike_id no nos brinda mayor informaci√≥n por s√≠ solo. Sin otros datos sobre la bicicleta, como por ejemplo modelo, tracci√≥n, o tama√±o, no sirve m√°s que para ver las m√°s usadas sin posibilitar alg√∫n analisis. Ser√≠a interesante analizar que caracter√≠sticas de la bicicleta importan m√°s seg√∫n condiciones clim√°ticas, que modelos son los m√°s usados.

```python
trips = trips[(trips.duration >=300) & (trips.duration <= 86400)]
trips['date'] = trips.start_date
trips['trip_id'] = trips.id

trips = trips[['trip_id','duration','date', 'season',
               'start_station_name','start_station_id','start_time','start_hour', 'day_of_week',
               'end_time','end_hour','end_station_name','end_station_id', 'subscription_type']]

trips.head(3)
```

```python
print trips.shape
```

```python
trips_total_count = trips['trip_id'].count()
print(trips_total_count)
```

En ambos gr√°ficos podemos observar como domina el verano en cuanto a duraci√≥n y cantidad de viajes. En particular, la duraci√≥n promedio de los viajes alcanza casi los 1250 segundos, es decir, 20 minutos.

Sin embargo, s√≠ hay clara diferencia en las duraciones de viajes. Los picos de 20 minutos como ya dijimos se dan en verano, pero decrecen rapidamente hacia el oto√±o, y m√°s a un en invierno. Recordando que la cantidad de viajes no variaba tanto, podemos apoyarnos m√°s sobre la teor√≠a de los viajes cortos rutinarios que realiza la gente durante todo el a√±o. Finalmente y contrario a nuestra intuici√≥n, en primavera se observan viajes m√°s cortos que en oto√±o. Esto debe deberse a otras variables clim√°ticas como lluvias, h√∫medad, adem√°s de la temperatura, que analizaremos m√°s adelante.
