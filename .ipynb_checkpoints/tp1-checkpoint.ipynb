## Analizando trip.csv

```python
trips = pd.read_csv('data/trip.csv',
                    parse_dates=['start_date', 'end_date'], infer_datetime_format=True,
                    low_memory=False)
```

```python
print trips.shape
trips.head(2)
```

```python
for i in trips.columns.values: print i
```

```python
trips['season'] = trips.start_date.dt.month

trips.season = trips.season.replace(1, 'winter')
trips.season = trips.season.replace(2, 'winter')
trips.season = trips.season.replace(3, 'winter')
trips.season = trips.season.replace(4, 'spring')
trips.season = trips.season.replace(5, 'spring')
trips.season = trips.season.replace(6, 'spring')
trips.season = trips.season.replace(7, 'summer')
trips.season = trips.season.replace(8, 'summer')
trips.season = trips.season.replace(9, 'summer')
trips.season = trips.season.replace(10, 'autumn')
trips.season = trips.season.replace(11, 'autumn')
trips.season = trips.season.replace(12, 'autumn')
```

```python
#separamos la fecha de las hora
trips['start_time'] = trips.start_date.dt.time
trips['start_hour'] = trips.start_date.dt.hour
trips['day_of_week'] = trips.start_date.dt.dayofweek
trips['start_date'] = trips.start_date.dt.date

trips['end_time'] = trips.end_date.dt.time
trips['end_hour'] = trips.end_date.dt.hour
trips['end_date'] = trips.end_date.dt.date

trips.head(2)
```

```python
trips.duration.describe()
```

```python
print trips.duration.max()
print trips.duration.min()
```

```python
print list(set(trips.subscription_type.values))
```

```python
# número de valores distintos que puede tomar zipcode (demasiados para print)
print len(list(set(trips.zip_code.values)))
```


Debido al alto número de trips con duraciones muy chicas, de pocos minutos, teorizamos que estos "trips" no son en verdad alquileres si no bicicletas retiradas para reparación o control. Asimismo podemos observar picos de "alquileres" en horas muy extrañas del día, que no por casualidad coinciden con los viajes de poca duración. Concluimos entonces que estos registros en el set de datos no corresponden a viajes reales.

Por lo tanto filtraremos los registros con duraciones menores a 5 minutos (300 segundos). A su vez, la misma página aclara que los viajes no puede durar más de 24hrs (86400 segundos) pues entonces se considera la bicicleta como 'robada', por lo que filtramos este tipo de viajes también.
Teniendo en cuenta también este dato, reemplazamos las columnas start_date y end_date por una única date correspondiente al día de comienzo del viaje.

Los zip_codes no parecen brindar información clara. Son demasiados valores y no corresponden con los zip codes de las 5 ciudades de weather y station.

Finalmente, bike_id no nos brinda mayor información por sí solo. Sin otros datos sobre la bicicleta, como por ejemplo modelo, tracción, o tamaño, no sirve más que para ver las más usadas sin posibilitar algún analisis. Sería interesante analizar que características de la bicicleta importan más según condiciones climáticas, que modelos son los más usados.

```python
trips = trips[(trips.duration >=300) & (trips.duration <= 86400)]
trips['date'] = trips.start_date
trips['trip_id'] = trips.id

trips = trips[['trip_id','duration','date', 'season',
               'start_station_name','start_station_id','start_time','start_hour', 'day_of_week',
               'end_time','end_hour','end_station_name','end_station_id', 'subscription_type']]

trips.head(3)
```

```python
print trips.shape
```

```python
trips_total_count = trips['trip_id'].count()
print(trips_total_count)
```

# Análisis por temporada

### Cantidad

```python
trips_season = pd.DataFrame({'trip_count' : trips.groupby(['season'])['trip_id'].count()}).reset_index()
trips_season
```

```python
fig = plt.figure(figsize=(10,4));

trips_season.set_index(['season'])['trip_count'].plot(kind='area')
```

### Duración

```python
trips.groupby('season').mean()['duration'].plot(figsize=(14,4));
```

En ambos gráficos podemos observar como domina el verano en cuanto a duración y cantidad de viajes. En particular, la duración promedio de los viajes alcanza casi los 1250 segundos, es decir, 20 minutos.

Por otro lado se puede ver que la cantidad de viajes no es tan superior a las otras estaciones. La cantidad de viajes del semestre frío (otoño e invierno) bastante parecida, incrementando hacia la primavera y llegando al pico en verano. La estación del año no parece influenciar mucho en la cantidad de viajes. Esto puede ser por la costumbre de la mayoría de usuarios de usar las bicicletas todo el año, para viajes rutinarios como trasladarse algún lado en particular como el trabajo.

Sin embargo, sí hay clara diferencia en las duraciones de viajes. Los picos de 20 minutos como ya dijimos se dan en verano, pero decrecen rapidamente hacia el otoño, y más a un en invierno. Recordando que la cantidad de viajes no variaba tanto, podemos apoyarnos más sobre la teoría de los viajes cortos rutinarios que realiza la gente durante todo el año. Finalmente y contrario a nuestra intuición, en primavera se observan viajes más cortos que en otoño. Esto debe deberse a otras variables climáticas como lluvias, húmedad, además de la temperatura, que analizaremos más adelante.

# Duración por hora

```python
trips.plot.scatter('duration','start_hour',alpha=0.25,figsize=(10,6));
plt.xlim(0,100000);
plt.ylim(0,24);
```

```python
trips_duration = pd.DataFrame({'duration_mean' : trips.groupby(['start_hour','end_hour'])['duration'].mean()}).reset_index()
trips_duration = trips_duration.sort_values(by='duration_mean', ascending=False)

duration = trips_duration.pivot("start_hour", "end_hour", "duration_mean")
fig, ax = plt.subplots(figsize=(16,5));
ax = sns.heatmap(duration, cmap = "OrRd", annot=False, linewidths=.5)
```

En estos gráficos podemos comparar los viajes y su duracipón según la hora del día que se registran. Podemos observar que los viajes más largos se concentran a partir de las 8 de la mañana, decreciendo hacia la tarde. Resaltamos que en las horas pico (6-8 y 17-20) se registran viajes cortos, por lo que es más que probable que los usuarios estén yendo o volviendo de trabajar.

En estos gráficos podemos observar claramente la diferencia en comportamiento entre clientes del tipo casuales o suscriptores. En particular observamos en un gráfico como los suscriptores registran muchos más viajes y están repartidos en todo el año, aunque con una pequeña baja en aquellos meses de otoño e invierno, seguramente debido al mayor frío. En el caso de los clientes, se registran mucha menor cantidad de viajes y también hay más en verano y primavera, esto se puede deber a que sean viajes mayormente recreativos. 
En cuanto a la concentración en los días, en los subscriptores hay una clara diferencia, ya que hay mayor concentración en los días laborales, más que los fines de semana. Para los clientes, al contrario, hay un leve aumento en el fin de semana, debido a lo ya mencionado del tipo de viaje que hacen los clientes. 
En el siguiente se puede ver que además de la gran cantidad de viajes de los suscriptores, estos se concentran en las horas pico, mientras que para los clientes hay un leve aumento, pero se debe a que son las horas del día, donde la gente suele estar despierta.

Las diferencias claras aplican también a la duracion. Podemos observar que los viajes de suscriptores son de menor duración que los de los clientes, tanto a lo largo del año como de la semana, e incluso según la hora del día. Esto se debe también al tipo de viajes que hacen cada uno.

Concluímos así que los suscriptores son aquellos que usan las bicicletas de forma rutinaria para trasladarse en la ciudad, probablemente a lugares de trabajo. Por lo que tiene sentido que se suscriban al sistema. Por el otro lado, los clientes no subscriptos son aquellos que usan ocasionalmente la bicicleta, seguramente más para recreación.
